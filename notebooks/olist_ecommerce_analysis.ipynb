## Merge datasets

```python
# Merge orders + customers
df = orders.merge(customers, on="customer_id", how="left")

# Merge with order_items
df = df.merge(order_items, on="order_id", how="left")

# Merge with products
df = df.merge(products, on="product_id", how="left")

# Merge with payments
df = df.merge(order_payments, on="order_id", how="left")

# Merge with reviews
df = df.merge(order_reviews, on="order_id", how="left")

# Merge with sellers
df = df.merge(sellers, on="seller_id", how="left")
df.head()
```

```python
df.info()
```

## Change datatype to datetime and rename

```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'], errors='coerce').dt.date
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'], errors='coerce').dt.date
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'], errors='coerce').dt.date
```

```python
df.rename(columns={
    'order_purchase_timestamp': 'purchase_date',
    'order_delivered_customer_date': 'delivered_date',
    'order_estimated_delivery_date': 'estimated_delivery_date'
}, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df.rename(columns = {'review_comment_message': 'review_message'}, inplace=True)
```

## Drop rows with critical NaNs

```python
df = df.dropna(subset=["purchase_date", "delivered_date", "price", "payment_value"])
```

```python
df.isnull().sum()
```

## Mark missing reviews

```python
df.loc[:, "review_message"] = df["review_message"].fillna("No review")
```

```python
df.loc[:, "product_category_name"] = df["product_category_name"].fillna("unknown_category")
```

```python
df.loc[:, "review_score"] = df["review_score"].fillna(-1)
```

## Drop unnecessary columns from all CSV's after merging

```python
df.drop(['customer_id', 'seller_id', 'seller_city', 'seller_state', 'product_id', 'review_id'], inplace=True, axis=1)
df.head()
```

## Feature Engineering

```python
df.loc[:, "revenue"] = df["price"] + df["freight_value"]
```

```python
df['purchase_date'] = pd.to_datetime(df['purchase_date'], errors='coerce')
df['delivered_date'] = pd.to_datetime(df['delivered_date'], errors='coerce')
df['estimated_delivery_date'] = pd.to_datetime(df['estimated_delivery_date'], errors='coerce')
```

```python
df.info()
```

```python
df.loc[:, "delivery_days"] = (
    df['delivered_date'] - df['purchase_date']
).dt.days
```

```python
df['is_late'] = df['delivery_days'] > (df['estimated_delivery_date'] - df['purchase_date']).dt.days
df['late_status'] = df['is_late'].map({True: 'Late', False: 'On Time'})
```

```python
df.drop('is_late', inplace=True, axis=1)
```

```python
# df.loc[:, "is_repeat_customer"] = df["customer_unique_id"].duplicated(keep="first")
order_counts = df.groupby("customer_unique_id")["order_id"].transform("count")

# Create customer_type column in df directly
df["customer_type"] = order_counts.apply(
    lambda x: "Repeat Customer" if x > 1 else "One-time Customer"
)
```

```python
!pip install deep_translator
from deep_translator import GoogleTranslator, exceptions
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm import tqdm
```

```python
unique_texts = df['review_message'].dropna().unique()

translator = GoogleTranslator(source='pt', target='en')

def translate_safe(text):
    try:
        return text, translator.translate(text)
    except exceptions.TranslationNotFound:
        return text, "Unknown"
    except Exception:
        return text, "Unknown"

translation_map = {}

with ThreadPoolExecutor(max_workers=10) as executor:
    futures = [executor.submit(translate_safe, txt) for txt in unique_texts]
    for future in tqdm(as_completed(futures), total=len(futures), desc="Translating"):
        original, translated = future.result()
        translation_map[original] = translated

df['review_message_en'] = df['review_message'].map(translation_map)
```
